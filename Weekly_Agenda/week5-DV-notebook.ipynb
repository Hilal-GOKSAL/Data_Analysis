```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
print(sns.get_dataset_names())
```

```python
mpg = sns.load_dataset("mpg")
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
#describe all features, both numerical and categorical
titanic.describe(include = "all")
```

```python
# get number of null values per columns
titanic.isnull().sum()
```

```python
# we can see all missing value in heatmap
sns.heatmap(titanic.isnull(), cbar=True);
```

```python
# we can see numeric column's corr in heatmap
sns.heatmap(titanic.corr(), annot=True);
```
