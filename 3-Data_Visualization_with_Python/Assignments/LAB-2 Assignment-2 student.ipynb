## 2: Plot 12 most dangerous cities

```python
a= kill.city.value_counts().head(12)
a
```

```python
# with seaborn barplot
fig, ax = plt.subplots(figsize=(8,5))
sns.barplot(x = a.index, y = a.values)
ax.set_xticklabels(a.index, rotation =45)
ax.set_title("Most Dangereous City", color = "b",fontsize=16)
plt.show()
```

Signature:
sns.countplot(
    *,
    x=None,
    y=None,
    hue=None,
    data=None,
    order=None,
    hue_order=None,
    orient=None,
    color=None,
    palette=None,
    saturation=0.75,
    dodge=True,
    ax=None,
    **kwargs,
)
Docstring:
Show the counts of observations in each categorical bin using bars.

```python
# with seaborn countplot
fig, ax = plt.subplots(figsize=(8,5))
sns.countplot(x = "city", data=kill, order= a.index)
ax.set_xticklabels(a.index, rotation =45)
ax.set_title("Most Dangereous City", color = "b",fontsize=16)
plt.show()
```

## 3: Plot 20 most dangerous states

```python
b= kill.state.value_counts().head(20)
b
```

```python
# with seaborn barplot

fig, ax = plt.subplots(figsize=(8,5))

sns.barplot(x = b.index, y = b.values, data=kill)

ax.set_xticklabels(b.index, rotation =45)
ax.set_title("Most dangereous states", color = "b",fontsize=16)
plt.show()
```

```python
# with seaborn countplot
fig, ax = plt.subplots(figsize=(8,5))
sns.countplot(x = "state", data=kill, order= b.index)


ax.set_xticklabels(b.index, rotation =45)
ax.set_title("Most dangereous states", color = "b",fontsize=16)

plt.show()
```
